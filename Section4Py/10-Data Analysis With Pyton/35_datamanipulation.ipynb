### Data manipulation Amd Analysis with pandas

```python
import pandas as pd
```

```python
df = pd.read_csv('data.csv')
```

```python
df.head(5)
```

```python
df.dtypes
```

```python
## handling missing values
# df.isnull().any(axis=1)
df.isnull().any()
```

```python
df.isnull().sum()
```

```python
df_filled = df.fillna(0)
```

```python
## filling missing values with the mean of the column

df['Sales_fillNA']= df['Sales'].fillna(df['Sales'].mean())
df.head(12)
```

```python
df.dtypes
```

```python
# Renaming Columns
df= df.rename(columns={'Sale Date':'Sales Date'})
df.head(5)
```

```python
## change datatype
df['Value_new']=df['Value'].fillna(df['Value'].mean()).astype(int)  ## fillna() is use to fill the blank space
df.head(10)
```

```python
df['New Value']=df['Value'].apply(lambda x: x*2)
df.head()
```

```python
## Data Aggregating and grouping
df.head()
```

```python
## grouped a column with other column details
grouped_mean = df.groupby('Product')['Value'].mean()
print(grouped_mean)
```

```python
## grouping two column with another one
grouping = df.groupby(['Product','Region'])['Sales'].sum()
print(grouping)
```

```python
## aggregating multiple functions
group_agg = df.groupby('Region')['Value'].agg(['mean','sum','count'])
print(group_agg)
```

```python
### merging and joing dataframes
# create sample dataframes

df1 = pd.DataFrame({'Key':['A','B','C'], 'Value1':[1,2,3]})
df2 = pd.DataFrame({'Key':['A','B','D'], 'Value1':[4,5,6]})
```

```python
df1
```

```python
df2
```

```python
## Merge Dataframes on the 'Key Columns'
pd.merge(df1,df2,on='Key',how='inner')
```

```python
pd.merge(df1,df2,on='Key',how='outer')
```

```python
pd.merge(df1,df2,on='Key',how='left')
```

```python
pd.merge(df1,df2,on='Key',how='right')
```

```python
pd.merge(df1,df2,how='cross')
```
